```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV file
data = pd.read_csv("first_validation_accuracy.csv")

# Find the index of the highest value
max_index = data["Validation Accuracy"].idxmax()
max_epoch = data["Epoch"][max_index]
max_value = data["Validation Accuracy"][max_index]

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(
    data["Epoch"],
    data["Validation Accuracy"],
    linestyle="-",
    color="b",
    label="Validation Accuracy",
)  # Line without markers

# Add a marker at the highest point
plt.scatter(
    max_epoch, max_value, color="#FF8C00", label="Highest Point"
)  # Deep orange color

# Annotate the highest point
plt.annotate(
    f"{max_value:.1f}%",
    (max_epoch, max_value),
    textcoords="offset points",
    xytext=(0, 10),  # Adjust the position of the annotation
    ha="center",
    fontsize=12,
    color="#FF8C00",
)

# Set the chart title and axis labels
plt.title("First Attack via API", fontsize=16)  # Updated title
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("Validation Accuracy (%)", fontsize=14)

# Set the Y-axis range and ticks
plt.ylim(0, 100)
plt.yticks(range(0, 101, 15))  # Y-axis ticks with intervals of 15

# Add legend
plt.legend(fontsize=12)

# Save the chart as a PNG image
plt.savefig("First_plot.png", dpi=300)  # Save the image with the updated file name
plt.show()
```

```python
# Read CSV files
validation_file = "first_validation_accuracy.csv"
training_file = "first_training_result.csv"

validation_data = pd.read_csv(validation_file)
training_data = pd.read_csv(training_file)

# Extract relevant columns
epochs_validation = validation_data["Epoch"]
validation_accuracy = validation_data["Validation Accuracy"]

epochs_training = training_data["Epoch"]
training_accuracy = training_data["Validation accuracy"] * 100  # For training accuracy

# Plot the line chart for both Validation and Training Accuracy
plt.figure(figsize=(10, 6))

# Plot Training Accuracy (blue, shorter line)
plt.plot(
    epochs_training,
    training_accuracy,
    color="b",  # Blue for Training Accuracy
    linestyle="-",
    linewidth=2,
    label="Training",  # Training comes first
)

# Plot Validation Accuracy (red, longer line)
plt.plot(
    epochs_validation,
    validation_accuracy,
    color="r",  # Red for Validation Accuracy
    linestyle="-",
    linewidth=2,
    label="Duplicated",  # Validation comes second
)

# Set chart title and axis labels
plt.title("Training Model Accuracy vs. Duplicated Model Accuracy", fontsize=14)
plt.xlabel("Epoch", fontsize=12)
plt.ylabel("Accuracy (%)", fontsize=12)

# Set Y-axis range and ticks
plt.ylim(0, 100)  # Set Y-axis range from 0 to 100
plt.yticks(range(0, 101, 10))  # Set Y-axis ticks at intervals of 10

# Remove grid lines
plt.grid(False)

# Display legend with correct order
plt.legend(
    loc="upper left",  # Legend position
    fontsize=10,  # Font size
)

# Save the chart
plt.savefig("First_attack.png", dpi=300)  # Save the image with the updated file name

# Show the chart
plt.show()
```
